```python
# now let's try to count all teh words in the text file in our spark context
rdd1=sc.textFile("/Users/carbon/Desktop/Semester-4/Big Data Analytics/Code/Lab/text2.txt")

def splitter(line):
    return line.strip().split()

def mapper(word):
    return (word,1)

rdd1.flatMap(splitter).map(mapper).collect()
```

```python
# defining areducer function to reduce by key
def reducer(x,y):
    return x+y

rdd1.flatMap(splitter).map(mapper).reduceByKey(reducer).collect()
```

```python
# practicing a filter function, which keeps elements where filter function is true
def filter4(word):
    if(len(word)>4):
        return True
    else:
        return False

rdd1.flatMap(splitter).filter(filter4).collect()
rdd1.flatMap(splitter).map(mapper).take(1)
```

```python
# practicing coalesce function
rdd2=sc.parallelize(range(9),9)
rdd2.coalesce(5).glom().collect()
```

```python
# iterating through key value pairs in loop fashion
rdd2=rdd1.flatMap(splitter).map(mapper)

for k,v in rdd2.groupByKey().collect():
    print(f"{k}, {len(list(v))}")
```
