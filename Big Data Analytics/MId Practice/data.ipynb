```python
data=pd.read_csv('Customer Data.csv')
```

```python
data=data.dropna()
data=data.drop('Gender',axis=1)
data.head()
```

```python
data=pd.DataFrame(StandardScaler().fit_transform(data))
```

```python
# data=pd.DataFrame(whiten(data, check_finite=True),index=None)
data.head()
```

```python
pca=PCA(n_components=2)
pca=pd.DataFrame(pca.fit_transform(data),columns=[0,1])
```

```python
# coode for elbow curve
wcss=[]
for i in range(1,7):
    kmeans=KMeans(i)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
sns.lineplot(range(1,7),wcss);
```

```python
kmeans=KMeans(n_clusters=5)
kmeans.fit_predict(data)
temp=kmeans.labels_
pca1=PCA(n_components=2)
centers=pd.DataFrame(pca1.fit_transform(kmeans.cluster_centers_),columns=[0,1])
sns.scatterplot(data=pca,x=0,y=1,hue=temp)
plt.scatter(x=centers[0],y=centers[1],marker='x',c='black');
```

```python
# implementing tsne
# tsne=TSNE(n_components=2,verbose=1,random_state=123)
# temp=tsne.fit_transform(data)
kmeans=KMeans(n_clusters=5)
kmeans.fit_predict(data)
labels=kmeans.labels_
temp=pd.DataFrame(temp,columns=[0,1])
sns.scatterplot(data=temp,x=0,y=1,hue=labels)
```

```python
# now we can move on to agglomerative clustering
distance_matrix=linkage(pca[[0,1]],method='single',metric='euclidean')
```

```python
dn=dendrogram(distance_matrix)
```

```python
prediction=fcluster(distance_matrix,20,criterion='maxclust')
```

```python
sns.scatterplot(data=pca,x=0,y=1,hue=prediction)
```

```python
data3,y1 = make_circles(n_samples=500, factor=0.7, noise=0.04)
data3_t,y2=make_circles(n_samples=613, factor=0.3, noise=0.04)
data3 = pd.DataFrame(np.vstack([data3,data3_t]),columns=[0,1])
y2[y2==1]=2
data3['output'] = np.hstack([y1,y2])
data3=pd.DataFrame(StandardScaler().fit_transform(data3))
sns.scatterplot(data=data3,x=0,y=1,s=20);
```

```python
neighbors = NearestNeighbors(n_neighbors=100)
neighbors_fit = neighbors.fit(data3)
distances, indices = neighbors_fit.kneighbors(data3)
distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.plot(distances)
```

```python
dbscan=DBSCAN(eps=0.2)
dbscan.fit_predict(data3)
sns.scatterplot(data=data3, x=0,y=1,hue=dbscan.labels_)
```
